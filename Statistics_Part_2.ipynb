# Statistics Part 2

#Q1. What is hypothesis testing in statistics?
-> Hypothesis testing is a statistical method used to make inferences about a population based on sample data. It involves testing an assumption (hypothesis) about a population parameter using statistical evidence.

#Q2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
-> The null hypothesis ($H_0$) states that there is no effect or no difference in a given context. The alternative hypothesis ($H_A$) suggests that there is an effect or difference. Hypothesis testing aims to determine wheather to reject $H_0$ in favor of $ H_A $.

#Q3. What is the significance level in hypothesis testing, and why is it important ?
-> The significance level(α) is the probability of rejecting the null hypothesis whwn it is actually true (Type 1 error). Common values are 0.05 or 0.01. It helps control the likelihood of making incorrect conclusions.

#Q4. What does a p-value represent in hypothesis testing?
-> A p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the observed value, assuming the null hypothesis is true. A low p - value (<α) suggests rejecting $H_0$.

#Q5. How do you interpret the p-value in hypothesis testing?
->
- If p ≤ α, reject $H_0$ ; there is strong evidence against it.
- If p > α, fail to reject $H_0$ ; there is not enough evidence to support $H_A$

#Q6. What are Type 1 and Type 2 error in hypothesis testing?
->
- Type 1 error: Rejecting $H_0$ when it is true (false positive).
- Type 2 error: Failing to reject $H_0$ when it is false (false negative).

#Q7. What is the difference between  a one- tailed and a two- tailed test in hypothesis testing?
->
- One - tailed test: Tests for an effect in one direction (e.g., greater than or less than).
- Two - tailed test: Tests for an effect in both directions(e.g., not equal to ).

#Q8. What is the Z-test, and when is it used in hypothesis testing?
-> The Z - test is used to compare sample and population means when the population variance is known and the sample size is large (n ≥ 30 ). It assumes a normal distribution.

#Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?


 $$  Z = \frac {X-\mu} {\sigma}  $$

Where :    
- X is the sample mean,
- u is the population mean,
- σ is the population standard deviation.
The Z-score represents how many standard deviations the sample mean is from the population mean.

#Q10. What is the T- distribution , and when should it be used instead of the normal distribution?
-> The T- distribution is a probability distribution used when the sample size is small (n < 30 ) and the population standard deviation is unknown. It has heavier tails compared to the normal distribution.

#Q11. What is the difference between a Z-test and T-test?
->
- Z-test is used when the population variance is known and the sample size is large.
- T - test is used when the population variance is unknown and the sample size is small.

#Q12. What is the T-test, and how is it used in hypothesis testing?
-> The t -test compares sample means to determine if they are significantly different . It is used for small sample sizes when the population standard deviation is unknown.

#Q13. What is the relationship between Z - test and T - test  in hypothesis testing ?
->
Both tests are used to compare means , but the Z- test is used for large samples with known variance, while the T - test is used for small samples with unknown variance.

#Q14. What is a confidence interval, and how is it used to interpret statistical results?
-> A confidence interval(CI) is a range of values that likely contains the true population parameter. It provides an estimate of uncertainty. A 95% CI means we are 95% confident that the  interval  contains the true value.

#Q15. What is the margin of error, and how does it affect the confidence interval?
-> The margin of error is the maximum expected difference between the true population parameter and the sample estimate. A larger margin of error results in a wider confidence interval, indicating more uncertainty.

#Q16. How is Bayes' Theorem used in statistics , and what is its significance?
-> Bayes' Theorem describes the probability of an event based on prior knowledge. It is used in updating probabilities as a new evidence is introduced , commonly applied in medical testing, spam filtering, and machine learning.

#Q17. What is the Chi-square distribution, and when is it used?
-> The Chi- square distribution is used for categorical data analysis, particularly in tests of independence and goodness of fit.

#Q18. What is the Chi- square goodness - of - fit test, and how is it applied?
-> The Chi- square goodness - of - fit test checks if a smaple distribution matches an expected distribution. It is applied to categorical data to determine if observed frequencies differ from expected frequencies.

#Q19. What is the F- distribution , and when is it used in hypothesis testing?
-> The F- distribution is used in variance analysis(ANOVA) and comparing multiple population variances. It is applied when testing hypothesis about variance ratios.

#Q20. What is an ANOVA test, and what are its assumptions?
-> ANOVA (Analysis of Variance) compares means of three or more groups to determine if at least one group is significantly different.
Assumptions:
- Independence of observations
- Normality of the data
- Homogeneity of variance (equal variances across groups).

#Q21. What are the different types of ANOVA tests ?
-> ANOVA (Analysis of variance) is used to compare means across multiple groups. The different types of ANOVA tests include:
1. One - Way ANOVA
 - Compares means of three or more independent groups based on one independent variable.
 - Example : Comparing test scores among students from three different schools.
  
2. Two - Way ANOVA
 - Compares means considering two independent variables.
 - Example: Examining the effect of both gender and study method on exam performance.

3. Repeated Measures ANOVA
 - Used when the same subjects are tested multiple times under different conditions.
 - Example : Measuring blood pressure of patients before, during, and after treatment.

4. MANOVA (Multivariate Analysis of Variance)
 - Extension of ANOVA that evaluates multiple dependent variables simultaneously.
 - Example : Studying the impact of diet on both weight loss and and cholesterol levels.

#Q22. What is the F- test , and how does it relate to hypothesis testing?
 -> The F- test is a statistical test used to compare variances or means among groups. It is commonly used in :    
1. ANOVA (Analysis of Variance )
- The F-test determines if the variance among group means is significantly different.

2. Comparing Two Variances
- Used to check if two populations have the same variance.

3. Regression Analysis
- Used to assess whether a regression model significantly explains the dependent variable.

Formula for the F- statistic
  $$ F = \frac{\text{Variance Between Groups }} {\text {Variance Within Groups}} $$

- A low F- value suggests that the groups are similar.

```python
#Q1 Write a Python program to perform a Z - test for comparing a sample mean to a known population mean and interpret the results.
import numpy as np
import scipy.stats as stats
# sample data
sample = np.array ([50, 52,53,49,51,54,55,48,47,50])
population_mean = 50
population_std = 5 # assuming known population standard deviation

# computer Z-score
sample_mean = np.mean(sample)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(len(sample)))
# Compute p - value
p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))
# Print results
print (f"Z-score: {z_score}")
print (f"P-value: {p_value}")


```

```python
#Q5 Create a Python function that calculates and visualize Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_errors():
  x = np.linspace(-4,4,100)
  y = stats.norm.pdf(x,0,1)

  plt.plot (x,y,label = "Null Distribution")

  plt.fill_between(x,y, where =(x> 1.96), color = 'red', alpha = 0.5, label = "Type 1 Error (alpha)")
  plt.fill_between(x,y, where = (x < -1.96), color = 'red', alpha = 0.5)

  shifted_y = stats.norm.pdf(x,1,1)
  plt.plot(x,shifted_y, label = "Alternative Distribution")
  plt.fill_between(x,shifted_y, where = (x < 1.96), color = 'blue', alpha = 0.5, label = "Type 2 Error (beta)")

  plt.legend()
  plt.show()
plot_errors()
```

```python
#Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats
def confidence_interval(data, confidence_level = 0.95):
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof = 1)
  n = len(data)
  margin_error = stats.t.ppf((1 + confidence_level) / 2, df = n - 1) * (sample_std / np.sqrt(n))
  lower_bound = sample_mean - margin_error
  upper_bound = sample_mean + margin_error
  return lower_bound, upper_bound

#Example usage
sample = np.random.normal(50,10,30)
ci = confidence_interval(sample)
print (f"95% Confidence Interval: {ci}")
```

```python
#Q10 Write a python program to calculate the margion of error for a given confidence level using sample data .
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence_level = 0.95):
  sample_mean = np.mean(data)
  sample_std = np.std(data, ddof = 1)
  n = len(data)
  margin_error = stats.t.ppf((1 + confidence_level) / 2, df = n - 1) * (sample_std / np.sqrt(n))
  return margin_error
# Example usage
sample = np.random.normal(50,10,30)
moe = margin_of_error(sample)
print (f"Margin of Error: {moe}")
```

```python
#Q11 Implement a Bayesian method using Bayes' Theorem in Python and explain the process.
def bayes_theorem (prior_A,prob_B_given_A, prob_B):
  return (prob_B_given_A * prior_A) / prob_B
# Example : Probability of having a disease given a positive test result
prior_disease = 0.01
prob_positive_given_disease = 0.95
prob_positive = 0.05

posterior = bayes_theorem(prior_disease, prob_positive_given_disease, prob_positive)
print (f"Posterior Probability: {posterior}")
```
